```python
import os
import random
import cv2
import albumentations as A
import splitfolders

# Cấu hình Augmentation (Chỉ dùng để sinh thêm ảnh thiếu)
transform = A.Compose([
    A.Rotate(limit=30, p=0.5),
    A.HorizontalFlip(p=0.5),
    A.RandomBrightnessContrast(p=0.2),
])

def balance_and_split(source_dir, temp_dir, target_count=150):
    classes = os.listdir(source_dir)
    for cls in classes:
        cls_path = os.path.join(source_dir, cls)
        images = os.listdir(cls_path)
        
        # 1. Gen thêm nếu thiếu
        current_count = len(images)
        if current_count < target_count:
            needed = target_count - current_count
            print(f"Lớp {cls} thiếu {needed} ảnh -> Đang sinh thêm...")
            for i in range(needed):
                img_name = random.choice(images)
                img = cv2.imread(os.path.join(cls_path, img_name))
                aug_img = transform(image=img)['image']
                cv2.imwrite(f"{cls_path}/aug_{i}.jpg", aug_img)
        
        # 2. Cắt bớt nếu thừa (Undersampling)
        elif current_count > 200:
            selected = random.sample(images, 200)
            # (Code move các ảnh thừa ra folder backup, chỉ giữ lại 200)

    # 3. Chia Train/Val/Test
    splitfolders.ratio(source_dir, output="Dataset_Final", seed=42, ratio=(.7, .15, .15))

# Kết quả: Folder "Dataset_Final" chuẩn chỉ để nạp vào model.
```
